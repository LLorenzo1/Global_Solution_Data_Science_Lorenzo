## **Descrição do Projeto**
Este projeto, desenvolvido como parte da disciplina de **Data Science** na **Global Solution**, realiza uma análise detalhada das emissões de CO2 no Brasil, comparando-as com o cenário global e de outras regiões, como EUA, OCDE e América Latina. Utilizamos técnicas de Machine Learning para prever emissões futuras e categorizar emissões per capita, além de integrar dados climáticos via API.

---

## **Objetivos Principais**
1. **Analisar Tendências:** Identificar padrões nas emissões de CO2 no Brasil entre 1990 e 2019.
2. **Comparar Cenários:** Comparar as emissões do Brasil com o mundo e regiões específicas.
3. **Prever e Classificar:** Implementar modelos de Machine Learning para prever emissões e categorizar emissões per capita.
4. **Enriquecer Dados:** Utilizar uma API para obter informações climáticas complementares ao estudo.

---

## **Principais Resultados**
- **Tendências de Emissões:**  
  As emissões totais do Brasil mostraram crescimento moderado, com emissões per capita relativamente estáveis devido ao uso significativo de fontes renováveis.

- **Comparações Internacionais:**  
  O Brasil emite significativamente menos CO2 per capita em comparação aos EUA e à média da OCDE.

- **Modelos de Machine Learning:**  
  - **Regressão Linear:** Previsões de emissões baseadas em PIB e consumo de energia apresentaram um alto coeficiente de determinação (\( R^2 \)).
  - **Classificação:** Categorias de emissões per capita (Baixo, Médio, Alto) foram classificadas com sucesso utilizando um modelo de Árvore de Decisão, com boa precisão e recall.

- **Enriquecimento com API:**  
  Dados climáticos do Brasil foram integrados ao projeto para análise complementar.

---

## **Metodologia**

### **1. Coleta e Preparação de Dados**
- Utilização das bases de dados do **Our World in Data** e **Banco Mundial** para informações de emissões, PIB, população e consumo de energia.
- Limpeza e transformação dos dados para garantir consistência e integridade.

### **2. Análise Exploratória**
- Visualização de tendências de emissões totais e per capita.
- Comparações com outras regiões (Mundo, EUA, OCDE e América Latina).

### **3. Modelagem de Machine Learning**
- **Regressão Linear:** Previsão de emissões totais com base em PIB e consumo de energia.
- **Classificação:** Categorias de emissões per capita (Baixo, Médio e Alto) utilizando Árvore de Decisão.

### **4. Enriquecimento de Dados**
- Integração de dados climáticos do Brasil via **API Open-Meteo**.
